## Using matplotlib to plot openCV images

```python
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

# Load our input image
image = cv2.imread('./images/glassbox.jpg')
cv2.imshow('Original', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

### openCV uses BGR format instead of RGB format. In order to be able to use matplotlib we need to change this format. 

```python
plt.imshow(image)
```

It looks interesting, as sky color should be some what similar to light blue. This is because of the BGR format.

There are two ways to solve this problem.
1. Use of cvtColor function along with the COLOR_BGR2RGB flag.
2. Use numpy to change the format.

#### Using cvtColor function

```python
BGRflags = [flag for flag in dir(cv2) if flag.startswith('COLOR_BGR') ]
print BGRflags
```

```python
imagePlt = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(imagePlt)
```

### Using numpy

```python
np.shape(image)
```

```python
# Convert RGB to BGR. We need exchange R and B leaving G as it is.
def convertImage(BGR_array):
    imageNp = np.zeros_like(BGR_array)
    imageNp[:,:, 0] = BGR_array[:,:, 2]
    imageNp[:,:, 1] = BGR_array[:,:, 1]
    imageNp[:,:, 2] = BGR_array[:,:, 0]
    return imageNp

plt.imshow(convertImage(image))
```

### Grayscaling Colored Images

Gray scale images only store intensity information within a given range of minimum and maximum. They are different from the bilevel images which consist of only black and white colors. In image processesing, colored images are often converted to grayscale images. The processesing time is reduced as we have to process only a 2D array instead of 3D. 

### Convertion

In this tutorial, we are going to convert a colred image to a grayscale in two ways.
1. By using cvtColor function.
2. By passing an argument to the imread function.

```python
# We use cvtColor, to convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale', gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(gray_image)
```
